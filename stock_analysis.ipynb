```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.embed import components
from bokeh.resources import CDN #this is used to generate dynamic link to the css and js so that when we upgrade Bokeh we may not get any problem
#CDN stands for content delivery network

start = datetime.datetime(2017, 7, 1)
end = datetime.datetime(2017, 7, 25)
df = data.DataReader(name = "GOOG", data_source="google", start = start, end = end)
df

output_notebook()

df

#key='9ed92c704005fb3ed918129f0a827323'
#df.index[df.Close> df.Open]

def inc_dec(c, o):
    if c>o:
        value = "Increase"
        
    elif c<o:
        value = "Decrease"
    return value
df["Status"]= [inc_dec(c,o) for c, o in zip(df.Close, df.Open)]

df["Middle"] = (df.Open+df.Close)/2
df["Height"] = abs(df.Open-df.Close)

p=figure(x_axis_type= 'datetime', width = 1000, height =300, responsive = True)
p.title.text = 'Candlestick Chart'
p.grid.grid_line_alpha=0.3
hours_12 = 12*60*60*1000
p.segment(df.index, df.High, df. index, df. Low)
p.rect(df.index[df.Status=="Increase"], df.Middle[df.Status=="Increase"],
       hours_12, df.Height[df.Status=="Increase"],
       fill_color ="#808080",line_color ="black")
p.rect(df.index[df.Status=="Decrease"], df.Middle[df.Status=="Decrease"],
       hours_12, df.Height[df.Status=="Decrease"],
       fill_color ="#FF69B4",line_color ="black")

# to generate the HTML and script of Bokeh chart calling the components function from Bokeh.embed
# here passing the figure object which is p, p stored the figure of the chart
#components(p) -this will print the resource of 

script1, div1, = components(p)
#print(script1)# it will print the java script only
#print(div1)  # it will print the html only
cdn_js = CDN.js_files
cdn_css = CDN.css_files

# below two lines are only to output a file locally and show is using show method
output_file("CS.html")
show(p)
```
